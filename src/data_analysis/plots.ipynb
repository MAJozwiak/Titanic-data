```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv(r"C:\Users\marta\PycharmProjects\Titanic-data\data\Titanic-Dataset.csv")
```

```python
data['Age'].fillna(data['Age'].mean(), inplace=True)
data['Age'] = data['Age'].round()
data["Age"]=data["Age"].astype(int)
data.drop("Cabin", axis=1, inplace=True)
```

```python
value_counts = data['Survived'].value_counts()
percentages = value_counts / value_counts.sum() * 100
gp=data["Survived"].value_counts().plot.bar(
title=('Number of People who Survived vs Died'),
ylabel=('Number of Survivors'),
xlabel=("Survival"),
color=("Black","Red")
)
gp.set_xticklabels(labels=["Died", "Survived"])


for index, value in value_counts.items():
    percentage = percentages[index]
    gp.text(index, value, f'{percentage:.2f}%', ha='center', va='bottom')
```

```python
jj=data.groupby(["Sex"]).agg({"Survived":"sum"}).plot.bar(
title=('Number of Men and Women Who Survived'),
ylabel=('Number of Survivors'),
xlabel=("Gender"),
color=("Red"))
jj.set_xticklabels(labels=["Woman", "Man"])


```

```python
gr=data.groupby(['Pclass','Survived'])[['Survived']].value_counts().unstack().plot.bar(
title=('Survival by a Passenger Class'),
ylabel=('Number of Survivors'),
xlabel=("Passenger class"),
color=("Black","Red")
)
gr.set_xticklabels(labels=["1st Class", "2nd Class","3rd Class"])
gr.legend(["Died","Survived"])
```

```python
gr=data.groupby(['Pclass','Survived','Sex'])[['Survived']].value_counts().unstack().plot.bar(
title=('Survival by Sex and Passenger Class'),
ylabel=('Number of Survivors'),
xlabel=("Passenger class"),
color=("Black","Red")
)

gr.set_xticklabels(labels=["1rd Class Non-Survived", "1nd Class Survived","2rd Class Non-Survived","2nd Class Survived","3rd Class Non-Survived","3nd Class Survived"])
gr.legend(["Women","Men"])
```

```python
ee=data.groupby(['Embarked','Survived'])[['Survived']].value_counts().unstack().plot.bar(
title=('Survival by Embarked'),
ylabel=('Number of Surivors'),
color=("Black","Red")
)
ee.legend(['Non-Survivors', 'Survivors'])
```
